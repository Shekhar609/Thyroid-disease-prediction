In the above data we got to know that most of the column are having values as t or f so it will be better to map 't' as 1 and 'f' as 0

```python
col_mapping = {'f': 0, 't': 1, 2: 2}

# Columns to iterate over
columns_to_iterate = ['on thyroxine', 'query on thyroxine','on antithyroid medication' , 'sick' , 
                      'pregnant' , 'thyroid surgery' , 'I131 treatment' , 'query hypothyroid' , 'query hyperthyroid', 
                      'lithium' , 'goitre' , 'tumor' , 'hypopituitary' , 'psych' , 'TSH measured' , 'T3 measured' , 'TT4 measured',
                     'T4U measured' , 'FTI measured' , 'TBG measured']

# Iterate over each row in the specified columns
for col_name in columns_to_iterate:
    for i, value in enumerate(data[col_name]):
        if value in col_mapping:
            data.loc[i, col_name] = col_mapping[value]
        else:
            data.loc[i, col_name] = None
```

After mapping the t and f we need to map the columns 'sex' , 'referral source' and 'binaryClass'

```python
col1_mapping = {'M': 0, 'F': 1, 2: 2}
# Iterate over each row in the 'Sex' column
for i, value in enumerate(data['sex']):
    if value in col1_mapping:  # Convert value to string before comparison
        data.loc[i, 'sex'] = col1_mapping[str(value)]
    else:
        data.loc[i, 'sex'] = None
```

```python
col2_mapping = {'SVHC' : 0 , 'other' : 1 , 'SVI' : 2 ,  'STMW' : 3 , 'SVHD' : 4}
# Iterate over each row in the 'referral source' column
for i, value in enumerate(data['referral source']):
    if value in col2_mapping:
        data.loc[i, 'referral source'] = col2_mapping[str(value)]
    else:
        data.loc[i, 'referral source'] = None
```

```python
col3_mapping = {'N': 0, 'P': 1, 2: 2}
# Iterate over each row in the 'binaryClass' column
for i, value in enumerate(data['binaryClass']):
    if value in col3_mapping:
        data.loc[i, 'binaryClass'] = col3_mapping[str(value)]
    else:
        data.loc[i, 'binaryClass'] = None
```

## Missing Value Treatment

### Method - 1 -> Using K - mean for null value replacement

```python
from sklearn.cluster import KMeans

data_withoutMD = data.copy()  # Create a copy of the original data

# Step 1: Convert the "sex" column to numeric (assuming it contains numeric values and "?" as a string)
data_withoutMD['sex'] = pd.to_numeric(data_withoutMD['sex'], errors='coerce')

# Step 2: Create a new column for K-means clustering with "sex" values
data_withoutMD['sex_cluster'] = data_withoutMD['sex']

# Step 3: Perform K-means clustering to replace "?" values
# Prepare the data for clustering (drop any non-numeric or missing values)
cluster_data = data_withoutMD[['sex_cluster']].dropna()

# Perform K-means clustering with 2 clusters (assuming male and female)
kmeans = KMeans(n_clusters=2 , n_init=10)
kmeans.fit(cluster_data)

# Get the cluster centers (representing the median values of each cluster)
cluster_centers = kmeans.cluster_centers_

# Convert cluster centers to integers after rounding
cluster_centers = np.round(cluster_centers).astype(int)

# Replace missing values with the cluster centers
data_withoutMD['sex'].fillna(cluster_centers[0, 0], inplace=True)

# Convert the "sex" column back to int and then to str
data_withoutMD['sex'] = data_withoutMD['sex'].astype(int).astype(str)

# Drop the intermediate "sex_cluster" column
data_withoutMD.drop(columns=['sex_cluster'], inplace=True)

data_withoutMD.head()
```

Here I have the missing value in the column which was present in '?' by the center of the cluster

```python
columns_to_convert = ['age', 'TT4', 'FTI']

# Convert columns to numeric with 'coerce' to handle non-numeric values
data_withoutMD[columns_to_convert] = data_withoutMD[columns_to_convert].apply(pd.to_numeric, errors='coerce')

# Calculate median for each column
column_medians = data_withoutMD[columns_to_convert].median()

# Replace NaN values with the median of each column
data_withoutMD[columns_to_convert] = data_withoutMD[columns_to_convert].fillna(column_medians)

# Perform K-means clustering for null value replacement
for col in columns_to_convert:
    # Prepare the data for clustering (drop any non-numeric or missing values)
    cluster_data = data_withoutMD[[col]].dropna()
    
    kmeans = KMeans(n_clusters=2 , n_init=10)
    kmeans.fit(cluster_data)
    
    cluster_centers = kmeans.cluster_centers_
    
    # Convert cluster centers to string format
    cluster_centers_str = cluster_centers.astype(str)
    
    # Replace NaN values with the cluster centers
    data_withoutMD.loc[data_withoutMD[col].isnull(), col] = cluster_centers_str[0, 0]

# Convert columns back to object (string) type to represent null values with 'None'
data_withoutMD[columns_to_convert] = data_withoutMD[columns_to_convert].astype('object')

data_withoutMD.head()
```

In the above missing value treatment for col 'age', 'TT4', 'FTI' I have use median to replace the missing value


```python
columns_to_convert = ['TSH', 'T3', 'T4U']

# Convert columns to numeric with 'coerce' to handle non-numeric values
data_withoutMD[columns_to_convert] = data_withoutMD[columns_to_convert].apply(pd.to_numeric, errors='coerce')

# Calculate mean for each column
column_means = data_withoutMD[columns_to_convert].mean()

# Replace NaN values with the mean of each column
data_withoutMD[columns_to_convert] = data_withoutMD[columns_to_convert].fillna(column_means)

# Perform K-means clustering for null value replacement
for col in columns_to_convert:
    cluster_data = data_withoutMD[[col]].dropna()
    
    kmeans = KMeans(n_clusters=2 , n_init=10)
    kmeans.fit(cluster_data)
    
    # Get the cluster centers (representing the mean values of each cluster)
    cluster_centers = kmeans.cluster_centers_
    
    # Convert cluster centers to string format
    cluster_centers_str = cluster_centers.astype(str)
    
    # Replace NaN values with the cluster centers
    data_withoutMD.loc[data_withoutMD[col].isnull(), col] = cluster_centers_str[0, 0]

# Convert columns back to object (string) type to represent null values with 'None'
data_withoutMD[columns_to_convert] = data_withoutMD[columns_to_convert].astype('object')

data_withoutMD.head()
```

### Using adaboost algorithm on decision tree

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Create the base estimator (Decision Tree classifier)
base_estimator = DecisionTreeClassifier()

# Lists to store F1-Score and Accuracy values for different n_estimators
f1_scores = []
accuracies = []

# Range of n_estimators to test
n_estimators_range = np.arange(1, 100, 5)

for n_estimators in n_estimators_range:
    # Create the AdaBoost classifier with Decision Trees as base estimators and set n_estimators
    clf = AdaBoostClassifier(estimator=base_estimator, n_estimators=n_estimators, random_state=99)
    
    # Fit the model
    clf.fit(X_train, y_train)
    
    # Predict on the test set
    y_pred = clf.predict(X_test)
    
    # Calculate performance metrics
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    
    # Append F1-Score and Accuracy to lists
    f1_scores.append(f1)
    accuracies.append(accuracy)

    # Print performance metrics for each n_estimators value
    print(f"n_estimators: {n_estimators} - F1-Score: {f1:.4f}, Accuracy: {accuracy:.4f}")

# Plotting F1-Score and Accuracy versus the number of estimators
plt.figure(figsize=(10, 6))
plt.plot(n_estimators_range, f1_scores, label='F1-Score', marker='o')
plt.plot(n_estimators_range, accuracies, label='Accuracy', marker='o')
plt.xlabel('Number of Estimators (n_estimators)')
plt.ylabel('Score')
plt.title('F1-Score and Accuracy vs. Number of Estimators')
plt.legend()
plt.grid(True)
plt.show()
```
